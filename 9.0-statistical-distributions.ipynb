## Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t, chi2
```

## Normal Distribution

### Probability Density Function

```python
sample_space = np.arange(-5, 5, 0.001)

mean = 0
stdev = 1.0

pdf = norm.pdf(sample_space, mean, stdev)

plt.plot(sample_space, pdf)
plt.show();
```

### Cumulative Density Function

```python
sample_space = np.arange(-5, 5, 0.001)

mean = 0
stdev = 1.0

cdf = norm.cdf(sample_space, mean, stdev)

plt.plot(sample_space, cdf)
plt.show();
```

## Student's-t distribution

### Probability Density Function

```python
sample_space = np.arange(-5, 5, 0.001)

dof = len(sample_space) -1

pdf = t.pdf(sample_space, dof)

plt.plot(sample_space, pdf)
plt.show();
```

### Cumulative Density Function

```python
sample_space = np.arange(-5, 5, 0.001)

dof = len(sample_space) -1

cdf = t.cdf(sample_space, dof)

plt.plot(sample_space, cdf)
plt.show();
```

## Chi-squared distribution

### Probability Density Function

```python
sample_space = np.arange(0, 50, 0.01)

dof = 20

pdf = chi2.pdf(sample_space, dof)

plt.plot(sample_space, pdf)
plt.show();
```

### Cumulative Density Function

```python
sample_space = np.arange(0, 50, 0.01)

dof = 20

cdf = chi2.cdf(sample_space, dof)

plt.plot(sample_space, cdf)
plt.show();
```
