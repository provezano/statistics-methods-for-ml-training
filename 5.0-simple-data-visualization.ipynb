## Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line Plot

```python
x = [x*0.1 for x in range(100)]
y = np.sin(x)

plt.plot(x, y)
plt.show();
```

## Bar Chart

```python
np.random.seed(1)
x = ['red', 'green', 'blue']
y = [np.random.randint(0, 100) for i in range(3)]

plt.bar(x, y)
plt.show()
```

## Histogram

```python
np.random.seed(1)
x = [np.random.randn(1000)]
plt.hist(x, bins=50)
plt.show()
```

## Boxplot

```python
np.random.seed(1)
x = [i*np.random.randn(1000) for i in range(1,15,5)]
plt.boxplot(x)
plt.show()
```

### Scatterplot

```python
np.random.seed(1)
x = 20 * np.random.randn(1000) + 100
y = x + (10 * np.random.randn(1000) + 50)
plt.scatter(x, y)
plt.show()
```
