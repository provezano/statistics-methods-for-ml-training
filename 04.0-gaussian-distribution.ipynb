## Libraries

```python
import pylab
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, probplot, mode
```

## The Normal Distributution

### Example 1

```python
x_axis = np.arange(-3, 3, 0.001)
y_axis = norm.pdf(x_axis, 0, 1)

plt.plot(x_axis, y_axis)
plt.show();
```

### Example 2

```python
np.random.seed(1)
data = 5 * np.random.randn(10000) + 50
plt.hist(data)
plt.show()
```

```python
probplot(data, dist="norm", plot=pylab)
pylab.show()
```

### Example 3

```python
np.random.seed(1)
data = 5 * np.random.randn(10000) + 50
plt.hist(data, bins=100)
plt.show()
```

```python
probplot(data, dist="norm", plot=pylab)
pylab.show()
```

## Measures of Location

### Mean

```python
np.random.seed(1)
data = 5 * np.random.randn(10000) + 50
print(np.mean(data), np.average(data))
```

### Median

```python
np.random.seed(1)
data = 5 * np.random.randn(10000) + 50
print(np.median(data))
```

### Mode

```python
np.random.seed(1)
data = 5 * np.random.randint(1, 10, 10000) + 50
print(mode(data))
```

## Measures of Dispersion

### Variance

```python
x_axis = np.arange(-3, 3, 0.001)

y_axis_1 = norm.pdf(x_axis, 0, 0.5)
y_axis_2 = norm.pdf(x_axis, 0, 1)


plt.plot(x_axis, y_axis_1)
plt.plot(x_axis, y_axis_2)
plt.show();
```

```python
np.random.seed(1)
data = 5 * np.random.randn(10000) + 50
print(np.var(data))
```

### Standard Deviation

```python
np.random.seed(1)
data = 5 * np.random.randn(10000) + 50
print(np.std(data))
```
