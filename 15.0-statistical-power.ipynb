### Libraries

```python
import numpy as np
from statsmodels.stats.power import TTestIndPower
```

### Set up

```python
np.random.seed(1)
```

```python
effect = 0.8
alpha = 0.05
power = 0.8
```

### Power analysis

#### Computing sample size

```python
analysis = TTestIndPower()
result = analysis.solve_power(effect,
                              power=power,
                              nobs1=None,
                              ratio=1.0,
                              alpha=alpha)
print('Sample size: {}'.format(np.ceil(result)))
```

#### Power curves

```python
analysis = TTestIndPower()
result = analysis.plot_power(dep_var = 'nobs',
                            nobs = np.arange(5, 100),
                            effect_size = np.array([0.2, 0.5, 0.8]))
```
